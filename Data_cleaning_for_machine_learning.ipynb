```python
import pandas as pd
df = pd.read_csv('AfSBBB_BE_LOR_Strasse_Strassenverkehrsunfaelle_2021_Datensatz (1).csv', delimiter=';')
df
```

```python
df.dtypes
```

```python
#Changing column types

df['IstRad'] = df['IstRad'].astype(float)
df['IstPKW'] = df['IstPKW'].astype(float)
df['IstFuss'] = df['IstFuss'].astype(float)
df['IstKrad'] = df['IstKrad'].astype(float)
df['IstGkfz'] = df['IstGkfz'].astype(float)
df['IstSonstige'] = df['IstSonstige'].astype(float)
df['UKATEGORIE'] = df['UKATEGORIE'].astype(float)
df['UART'] = df['UART'].astype(float)
df['UTYP1'] = df['UTYP1'].astype(float)
df['ULICHTVERH'] = df['ULICHTVERH'].astype(float)
df['USTRZUSTAND'] = df['USTRZUSTAND'].astype(float)


df['XGCSWGS84'] = df['XGCSWGS84'].str.replace(',', '.').astype(float)
df['YGCSWGS84'] = df['YGCSWGS84'].str.replace(',', '.').astype(float)

df['UJAHR'] = df['UJAHR'].map(str)
df['UMONAT'] = df['UMONAT'].map(str)
df['USTUNDE'] = df['USTUNDE'].map(str)
df['UWOCHENTAG'] = df['UWOCHENTAG'].map(str)

df.dtypes
```

```python
df['UKATEGORIE'].value_counts() #Exploring target variable
```

```python
#Translating columns to English
df.rename(columns = {'OBJECTID':'id',
'LAND':'land',
'BEZ': 'district',
'LOR_ab_2021':'lor_ab_2021',
'UJAHR':'year',
'UMONAT':'month',
'USTUNDE':'hour',
'UWOCHENTAG':'day_of_week',
'UKATEGORIE':'accident_category',
'UART':'accident_details',
'UTYP1':'accident_type',
'ULICHTVERH':'light_condition',
'IstRad':'acc_with_bike',
'IstPKW':'acc_with_car',
'IstFuss':'acc_with_ped',
'IstKrad':'acc_with_motobike',
'IstGkfz':'acc_with_truck',
'IstSonstige':'acc_with_other',
'USTRZUSTAND':'road_condition',
'LINREFX':'LINREFX', 
'LINREFY':'LINREFY',
'XGCSWGS84':'longitude',
'YGCSWGS84':'latitude'}, inplace = True)
```

```python
df = df.dropna()
df=df.drop(columns=['lor_ab_2021', 'LINREFX', 'LINREFY'])
df
```

```python
df.to_excel("Berlin_accidents_clean.xlsx", index=False)
```
