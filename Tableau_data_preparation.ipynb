```python
import pandas as pd

df = pd.read_excel('Berlin_accidents_clean.xlsx')
df
```

```python
#Getting Months names
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
               7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

# use the map() method to replace the encoded months with their corresponding month names
df['month'] = df['month'].map(month_names)
df
```

```python
#Getting district names
district_names = {1: 'Mitte', 2: 'Friedrichshain-Kreuzberg', 3: 'Pankow', 4: 'Charlottenburg-Wilmersdorf', 5: 'Spandau', 6: 'Steglitz-Zehlendorf',
               7: 'Tempelhof-SchÃ¶neberg', 8: 'NeukÃ¶lln', 9: 'Treptow-KÃ¶penick', 10: 'Marzahn-Hellersdorf', 11: 'Lichtenberg', 12: 'Reinickendorf'}

# use the map() method to replace the encoded months with their corresponding month names
df['district'] = df['district'].map(district_names)
df
```

```python
#Getting day of the week
weekday_names = {1: 'Sunday', 2: 'Monday', 3: 'Tuesday', 4: 'Wednesday', 5: 'Thursday', 6: 'Friday',
               7: 'Saturday'}

# use the map() method to replace the encoded months with their corresponding month names
df['day_of_week'] = df['day_of_week'].map(weekday_names)
df
```

```python
#Translating severity
severity = {1: 'fatalities', 2: 'serious injuries', 3: 'minor injuries'}

# use the map() method to replace the encoded months with their corresponding month names
df['accident_category'] = df['accident_category'].map(severity)
df
```

```python
#Translating day light
light = {0: 'daylight', 1: 'twilight', 2: 'darkness'}

# use the map() method to replace the encoded months with their corresponding month names
df['light_condition'] = df['light_condition'].map(light)
df
```

```python
#Translating road condition
light = {0: 'dry', 1: 'wet/damp/slippery', 2: 'winter slippery'}

# use the map() method to replace the encoded months with their corresponding month names
df['road_condition'] = df['road_condition'].map(light)
df
```

```python
df.to_excel("Tableau_Berlin_accidents_2021.xlsx", index=False)
```
